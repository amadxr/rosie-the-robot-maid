## Function declaration

```python
def try_parse_int(s, base=10, val=None):
    try:
        return int(s, base)
    except (ValueError, TypeError):
        return val

def company_size_transformer(company_size):
    if "-" in (str)(company_size):
        company_size = company_size.split("-")[-1]
    
    if "+" in (str)(company_size):
        company_size = company_size.replace("+", "")
    
    return try_parse_int(company_size)

def company_size_limiter(company_size):
    if company_size is not None and company_size > company_size_limit:
        company_size = None
        
    return company_size
```

## User Interaction Phase

```python
limit = input("Type the company size limit for this list (ENTER if you do not want to limit the size)")
global company_size_limit
company_size_limit = try_parse_int(limit)
```

## List Processing Phase

```python
leads['Company Size'] = leads['Company Size'].apply(company_size_transformer)

if company_size_limit is not None:
    leads['Company Size'] = leads['Company Size'].apply(company_size_limiter)
    
leads = leads[leads['Company Size'].notnull()]
leads.head(30)
        
```
